```python
import torch
from torch import nn
import torch.nn.functional as F
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
```

Let's generate the Devil's Toy dataset

```python
# the code for generating is taken from the Stanford course:
# http://cs231n.github.io/neural-networks-case-study/#linear

N = 100
D = 2
K = 3
X = np.zeros((N * K, D))
y = np.zeros(N * K, dtype='uint8')

for j in range(K):
    ix = range(N * j,N * (j + 1))
    r = np.linspace(0.0, 1, N)
    t = np.linspace(j * 4, (j + 1) * 4,N) + np.random.randn(N) * 0.2 # theta
    X[ix] = np.c_[r * np.sin(t), r * np.cos(t)]
    y[ix] = j
```

```python
# Rendering magic, taken from again:
# http://cs231n.github.io/neural-networks-case-study/#linear

plt.figure(figsize=(10, 8))

plt.scatter(X[:, 0], X[:, 1], c=y, s=40, cmap=plt.cm.rainbow)

plt.title('Toy of ]:) ', fontsize=15)
plt.xlabel('$x$', fontsize=14)
plt.ylabel('$y$', fontsize=14)
plt.show();
```

```python
X = torch.FloatTensor(X)
y = torch.LongTensor(y)
```

Now we want to learn how to independently create heirs to nn.Module. Before that, we made neural networks using the nn.Sequential class, let's try to build the same network as in the last seminar, but on our own.

```python
#NETWORK FROM LAST SEMINAR

# D_in - dimension of the input (the number of features of the object);
# H - dimension of hidden layers;
# D_out - the dimension of the output layer (the essence is the number of classes)
D_in, H, D_out = 2, 100, 3

# Use the nn package to define our model and loss function.
two_layer_net = torch.nn.Sequential(
    torch.nn.Linear(D_in, H),
    torch.nn.ReLU(),
    torch.nn.Linear(H, D_out),
    torch.nn.Softmax()
)
```

What is a module and how does it work? Firstly, a module is such a building block for neurons, with the help of the module you can set any function differentiable by its parameter. Modules are applied in the same way as regular functions with the syntax
> module_instance(var1, var2)

This calls the forward function internally with the same arguments, and returns its output as the result of calling the module. Why is it so strange to wrap ordinary functions in modules?

* This makes it very convenient to keep track of the parameters that need to be changed. When we want to get all the parameters, we can simply recursively go through all the fields of the model, see which of the parameters are in themselves, and which are modules and contain the parameters inside, and then collect all this.

_For this reason, if you use any other modules inside your module, you need to put them simply in the class field, if this is a single module, and in the ** nn.ModuleList ** or ** nn.ModuleDict ** class, if you have a list or a dictionary of used modules. If in the model you have some own weight, then it is not enough to put the tensor in the class field, it must be wrapped in ** nn.Parameter, nn.ParameterList ** or ** nn.ParameterDict **, depending on what it is with you._

* Such an organization allows you to fairly painlessly extend PyTorch and write your own functions for it, which cannot be expressed by the composition of existing ones. This is rarely useful, so today we will not write our extension.

* Modularized code is an implementation of the encapsulation concept from object-oriented programming.

```python
# New Model

class MyModule(nn.Module):
    def __init__(self):
        super().__init__()
        self.f_lin = nn.Linear(D_in, H)
        self.s_lin = nn.Linear(H, D_out)
        
    def forward(self, X):
        X = F.relu(self.f_lin(X))
        return F.softmax(self.s_lin(X))
    
model = MyModule()
print(list(model.parameters()))
loss_fn = nn.CrossEntropyLoss()
optim = torch.optim.Adam(model.parameters(), 1e-1)
```

```python
losses = []
for t in range(100):
    # forward
    y_pred = model(X)

    # loss
    loss = loss_fn(y_pred, y)
    losses.append(loss.item())

    optim.zero_grad()
    loss.backward()
    optim.step()
    
plt.plot(losses)
```

Now let's see what happens if we don't put the layers used inside in self var

```python
# Model without parameters

class MyModule(nn.Module):
    def __init__(self):
        super().__init__()
        self.linear_layers = [nn.Linear(D_in, H), nn.Linear(H, D_out)]
        self.my_useless_bias = torch.ones(1, H, requires_grad=True)
        self.more_of_my_useless_biases = [
            torch.ones(1, H, requires_grad=True),
            torch.ones(1, H, requires_grad=True),
            torch.ones(1, H, requires_grad=True)
        ]
        
    def forward(self, X):
        X = F.relu(self.linear_layers[0](X))
        X += self.my_useless_bias
        return F.softmax(self.linear_layers[1](X))
    
model = MyModule()
list(model.parameters())
```

As we expected, there are no parameters. Let's fix this.

```python
# Corrected model with params

class MyModule(nn.Module):
    def __init__(self):
        super().__init__()
        self.linear_layers = nn.ModuleList([nn.Linear(D_in, H), nn.Linear(H, D_out)])
        self.my_useless_bias = nn.Parameter(torch.ones(1, H, requires_grad=True))
        self.more_of_my_useless_biases = nn.ParameterList([
            nn.Parameter(torch.ones(1, H, requires_grad=True)),
            nn.Parameter(torch.ones(1, H, requires_grad=True)),
            nn.Parameter(torch.ones(1, H, requires_grad=True))
        ])
        
    def forward(self, X):
        X = F.relu(self.linear_layers[0](X))
        X += self.my_useless_bias
        for b in self.more_of_my_useless_biases:
            X += b
        return F.softmax(self.linear_layers[1](X))
    
model = MyModule()
list(model.parameters())
```

As we wanted, all parameters appeared
